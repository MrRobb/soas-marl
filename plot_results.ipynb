```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
sns.set_theme()
```

```python
df1 = pd.read_csv('results/results1.csv')
df2 = pd.read_csv('results/results2.csv')
df3 = pd.read_csv('results/results3.csv')
df4 = pd.read_csv('results/results4.csv')
df5 = pd.read_csv('results/results5.csv')
df6 = pd.read_csv('results/results6.csv')
df7 = pd.read_csv('results/results7.csv')
df8 = pd.read_csv('results/results8.csv')
df9 = pd.read_csv('results/results9.csv')
df10 = pd.read_csv('results/results10.csv')
df11 = pd.read_csv('results/results11.csv')
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11], ignore_index=True)
df
```

```python
sns.relplot(data=df, x="Training", y="Testing", kind="line")
plt.yscale("log")
plt.tight_layout()
plt.savefig("plots/timesteps.png")
plt.show()
```

```python
sns.displot(data=df, x="Testing", kde=True, bins=8)
plt.yscale("log")
plt.tight_layout()
plt.savefig("plots/timestep_dist.png")
plt.show()
```

```python
slip1 = pd.read_csv('results/slip1.csv')
slip2 = pd.read_csv('results/slip2.csv')
slip3 = pd.read_csv('results/slip3.csv')
slip4 = pd.read_csv('results/slip4.csv')
slip5 = pd.read_csv('results/slip5.csv')
slip6 = pd.read_csv('results/slip6.csv')
slip7 = pd.read_csv('results/slip7.csv')
slip8 = pd.read_csv('results/slip8.csv')
slip9 = pd.read_csv('results/slip9.csv')
slip10 = pd.read_csv('results/slip10.csv')
df_slip = pd.concat([slip1, slip2, slip3, slip4, slip5, slip6, slip7, slip8, slip9, slip10], ignore_index=True)
df_slip
```

```python
sns.relplot(data=df_slip, x="Training", y="Testing", kind="line")
plt.yscale("log")
plt.tight_layout()
plt.savefig("plots/slip.png")
plt.show()
```

```python
df_concat = pd.concat([df, df_slip], ignore_index=True, axis=1)
df_concat
```

```python
fig, ax = plt.subplots()
sns.lineplot(data=df, x="Training", y="Testing", estimator=np.median, ax=ax)
sns.lineplot(data=df_slip, x="Training", y="Testing", estimator=np.median, ax=ax)

plt.yscale("log")
plt.tight_layout()

red_patch = mpatches.Patch(color=(0.2980392156862745, 0.4470588235294118, 0.6901960784313725), label='No slip')
black_patch = mpatches.Patch(color=(0.8666666666666667, 0.5176470588235295, 0.3215686274509804), label='slip_pr=0.02')
plt.legend(handles=[red_patch, black_patch])

plt.savefig("plots/slip.png")
plt.show()
```
